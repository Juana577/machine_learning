```
# Playing with image intensities
img_gray = cv2.imread( "/content/drive/MyDrive/machine_learning/Sem_01/lena.jpg"  )

for r in range(img_gray.shape[0]):  # Go through rows
    for c in range(img_gray.shape[1]):  # Go through cols
        if img_gray is None:
           print("No se pudo cargar la imagen") # Crear una condición simple
        img_gray[r, c] = 0  # Establecer el valor del píxel a 0 (negro)


plt.show()
```
